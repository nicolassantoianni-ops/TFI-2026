```python
import networkx # importar librerias, la primera es para el grafo, la segunda para el print.
import matplotlib.pyplot # esto segun google es la libreria que se usa para graficar. lo saque de ahi.

G = networkx.DiGraph() # crea un grafo dirigido
nodos = [1, 2, 3, 4, 5, 6, 7, 8] # añade los 8 nodos
G.add_nodes_from(nodos)
conexiones = [ # crea las conexiones
    (1, 2), (1, 3),
    (2, 3), (2, 4),
    (3, 4), (3, 5),
    (4, 5), (4, 6),
    (5, 6), (5, 7),
    (6, 7), (6, 8),
    (7, 8), (7, 1),
    (8, 1), (8, 2)
]
G.add_edges_from(conexiones) # agrega las flechas

pos = networkx.spiral_layout(G) # dibujo del grafo, colores, tamaños 
networkx.draw(G, pos, with_labels=True, node_color='darkorchid',
        node_size=1200, edge_color='tomato', arrowsize=15,
        font_weight='bold', font_color='chartreuse', font_size=18 ) # si bien esto no se pide, le da una vuelta, tambien busque en google las posibilidades.

matplotlib.pyplot.title("Red: 8 nodos, Conexion dirigida: 2 por nodo") # titular y mostrar el grafico
matplotlib.pyplot.show()
```
